```python
import matplotlib.pyplot as plt
```

```python
# time (t), carrying capacity (K), growth rate (r), and initial population (P0) as parameters.
# Logistic growth function
def logistic_growth(t, K, r, P0):
    return K / (1 + (K / P0 - 1) * np.exp(-r * t))

# Parameters for the logistic growth
K = 1000  # Carrying capacity or the maximum population
r = 0.1   # Growth rate
P0 = 10   # Initial population

# Generate time values from 0 to 100 (you can adjust this range)
t = np.linspace(0, 100, 100)

# Calculate population at each time point
population = logistic_growth(t, K, r, P0)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t, population, label='Logistic Growth')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Logistic Growth Model')
plt.axvline(x=50, color='r', linestyle='--', label='Inflection Point')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# import numpy as np
# import matplotlib.pyplot as plt

# Parameters for exponential growth
r = 0.1  # Growth rate
N0 = 1   # Initial population
t_max = 10  # Maximum time period

# Generate time values from 0 to t_max
t = np.linspace(0, t_max, 100)

# Calculate population at each time point using the exponential growth formula
population = N0 * np.exp(r * t)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t, population, label='Exponential Growth')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Exponential Growth Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# import numpy as np
# import matplotlib.pyplot as plt

# We define parameters for linear growth: the growth rate (slope), initial value (intercept), and the maximum time period (t_max).
# We generate a range of time values using NumPy's linspace().
# We calculate the quantity at each time point using the linear growth formula, which is Quantity(t) = slope * t + intercept.



# Parameters for linear growth
slope = 2  # Linear growth rate
intercept = 10  # Initial value
t_max = 10  # Maximum time period

# Generate time values from 0 to t_max
t = np.linspace(0, t_max, 100)

# Calculate the linear growth at each time point
growth = slope * t + intercept

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t, growth, label='Linear Growth')
plt.xlabel('Time')
plt.ylabel('Quantity')
plt.title('Linear Growth Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# import numpy as np
# import matplotlib.pyplot as plt

# We define parameters for logarithmic growth: the limiting value K, growth rate r, and the maximum time period t_max.
# We generate a range of time values using NumPy's linspace().
# We calculate the quantity at each time point using the logarithmic growth formula, which is Quantity(t) = K * (1 - e^(-r * t)).

# Parameters for logarithmic growth
K = 100  # Limiting value or maximum value
r = 0.1  # Growth rate
t_max = 50  # Maximum time period

# Generate time values from 0 to t_max
t = np.linspace(0, t_max, 100)

# Calculate the quantity at each time point using the logarithmic growth formula
quantity = K * (1 - np.exp(-r * t))

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t, quantity, label='Logarithmic Growth')
plt.xlabel('Time')
plt.ylabel('Quantity')
plt.title('Logarithmic Growth Model')
plt.legend()
plt.grid(True)
plt.show()
```
